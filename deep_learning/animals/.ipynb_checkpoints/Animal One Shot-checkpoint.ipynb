```python
def load_image(filename):
    img = Image.open(filename)
    img.load()
    img = img.resize((150,150), Image.ANTIALIAS) 
    img = img.convert('L')
    data = np.asarray(img)
    return np.array([[[x] for x in y] for y in data])
```

```python
image_dir = "Image"
file_list = os.listdir("Image")
images_data = np.ndarray((2410, 150, 150, 1))
labels = []
categoryToId = os.listdir("Image")
num = 0
for dir in file_list:
    images = os.listdir('/'.join([image_dir, dir]))
    for image_name in images:
        labels.append(categoryToId.index(dir))
        tempImage = load_image('/'.join([image_dir, dir, image_name]))
        images_data[num] = tempImage    
        num += 1
```

```python
images_data.shape
```

```python
startOfLast5 = labels.index(16)
X = np.array(images_data) / 255
Y = np.array(labels)
X_train = X[:startOfLast5]
Y_train = Y[:startOfLast5]
X_test = X[startOfLast5:]
Y_test = Y[startOfLast5:]
```

```python
input_shape = X.shape[1:]
print(input_shape)
```

```python
animal_indices = [np.where(Y_train == i)[0] for i in range(15)]
tr_pairs, tr_y = sm.create_pairs(X_train, animal_indices,15)
```

```python
animal_indices = [np.where(Y_test == i)[0] for i in range(16,20)]
te_pairs, te_y = sm.create_pairs(X_test, animal_indices, 4)
```

```python
def create_network():
    base_network = Sequential()
    base_network.add(Conv2D(64, kernel_size=(4, 4), strides=(4, 4),
                     activation='relu',
                     input_shape=input_shape))
    base_network.add(MaxPooling2D(pool_size=(2, 2)))
    base_network.add(Conv2D(32, kernel_size=(5, 5), strides=(5, 5),
                     activation='relu',
                     input_shape=input_shape))
    base_network.add(MaxPooling2D(pool_size=(2, 2)))
    #base_network.add(Dropout(0.5))
    base_network.add(Flatten())
    base_network.add(Dense(256, activation='tanh'))

    #base_network = Model(input, x)
    input_a = Input(shape=input_shape)
    input_b = Input(shape=input_shape)
    processed_a = base_network(input_a)
    processed_b = base_network(input_b)
    model = Model([input_a, input_b], distance)
    optimizer = Adam(0.0001)
    model.compile(loss=sm.contrastive_loss, optimizer=optimizer, metrics=[sm.accuracy])
    # This is from one run with ADAM 0.0001
    model.fit([tr_pairs[:, 0], tr_pairs[:, 1]], tr_y,
          batch_size=128,
          epochs=50,
          validation_data=([te_pairs[:, 0], te_pairs[:, 1]], te_y))
    return model
```

```python
base_network = Sequential()
base_network.add(Conv2D(32, kernel_size=(15, 15), strides=15,
                 activation='relu',
                 input_shape=input_shape))
base_network.add(BatchNormalization(axis=-1))
base_network.add(MaxPooling2D(pool_size=(2, 2)))
base_network.add(Dropout(0.25))
base_network.add(Conv2D(64, kernel_size=(5, 5), strides=5,
                 activation='relu',
                 input_shape=input_shape))
base_network.add(BatchNormalization(axis=-1))
base_network.add(MaxPooling2D(pool_size=(2, 2)))
base_network.add(Dropout(0.25))
base_network.add(Flatten())
base_network.add(Dense(512, activation='relu'))
base_network.add(BatchNormalization(axis=-1))
```
