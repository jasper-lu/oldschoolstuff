## Mini-Batch discussion
First of all, in general, it seems like smaller choices of mini-batch size give us better performance. Like we have established in the past, training performance for our neural networks for this task generally seem to depend on how quickly we can escape from the big early local minimum (at classifying all images to false).

Perhaps using mini-batches instead of the whole dataset at a time when running our gradient descent gives us enough noise to help us escape from local minima more quickly. In addition, running gradient descent with mini-batches technically acts like us running more than one iteration of gradient descent per true iteration, but just with a smaller dataset size. We already know from prior experiments that we can achieve pretty good performance on a small batch. For example, when we run experiments on gradient descent with 600 instances instead of the entire 60000, we already achieve decent performance on this classification problem.

Also, since we are running on so many different mini-batches at a time, I would argue we reduce our chances of over-fitting as well. We can see evidence for this by looking at how close training and test errors tend to be with this particular classification problem. Perhaps by "shaking up" the gradient descent by using different mini-batches for each iteration--all of which are supposed to have the same underlying distribution--we disallow the network from learning too much of the particulars of any one specific mini-batch.

So, mini-batch gradient descent has the potential to give us both quicker learning and less bias in our neural network with respect to the training data. This makes mini-batch gradient descent a great choice for working with training neural networks on data. The only disadvantage is that we would never reach our true global optimum. However, we do not really expect to reach that point when training a neural network, anyway--it would be far too expensive to train a neural network up to that point.

It still seems like higher batch sizes tend to lead to better end performance in general. However, I would need more experimentation to confirm this, but it seems like smaller batch sizes tend to break through the local minimum more quickly, perhaps because the descent ends up much more noisy.

## Momentum discussion
In the first trial, I did not scale our $\alpha$ with the momentum ($\beta$). In the second one, I scaled $\alpha$ by a factor of $(1 - \beta)$, to more closely reflect the way momentum gradient descent is supposed to work. Scaling $\alpha$ with $(1 - \beta)$ is supposed to give us the effect of calculating a running average for velocity.

We achieve much better results on our gradient descent when we keep $\alpha$ stable than when we scale it to beta. Note that the reason for this is because when we fix $\alpha$ to $1$ and then change the $\beta$, it is as if our "true" $\alpha$ is actually $\frac{\alpha}{1 - \beta}$. So, trial 1 could also be interpeted as representing varying choices of learning rate along with varying choices of momentum.

In trial 2, because we are scaling $\alpha$, I suspect we end up achieve sub-par performance (compared to trial 1) because scaling $\alpha$ according to $\beta$ gives us an $\alpha$ which is far too small to effectively learn from.

We see the effects of choosing a larger initial $\alpha$ in trial 3. We end up achieving very good performance for all choices of momentum. What this tells me is really that when we use momentum gradient descent, we should consider using a larger learning rate than average, due to potential scaling. Furthermore, across all of our trials, a learning rate of 0.9 has consistently given us the best results. Perhaps this should be our starting point when we try to perform stochastic gradient descent with momentum.

What I understand from running these experiments is that, in this case with varying our momentum parameter, just like in varying any other hyperparameter, we should be careful what we choose, lest we might end up with a subpar model compared to what we *could* have.

In any case, momentum gradient descent really seems to outperform the non-momentum based version of gradient descent. Based on empirical results, especially looking at how quickly certain choices of momentum get to a high classification correctness rate, I would argue that adding momentum to our gradient descent algorithm allows us to overcome local minima more easily than non-momentum gradient descent methods. We can see this by imagining a ball rolling down a bowl. Even when the ball is in the middle of the bowl, it will still have momentum from its descent propelling it forward. I imagine this is similar to what happens in momentum gradient descent to help us escape and avoid local minima.

The plots from trial 3 really illustrate this hypothesis. All neural networks trained in trial 3 break out of their local minima in below 20 iterations, which is much much quicker than in all of our experiments up until this point.

```python
import sklearn.datasets as sk
import numpy as np
from relu_net import NeuralNet
```

```python
def normalize(X):
    count = 0
    for i, x in enumerate(X[1]):
        if x > 0:
            X[1][i] = 1
        else:
            X[1][i] = 0
    return X
```

```python
testX, testY = sk.make_moons()
trainX, trainY = sk.make_moons(n_samples = 400)
testX = normalize(testX.T)
trainX = normalize(trainX.T)
testY = testY.reshape(1, 100)
trainY = trainY.reshape(1, 400)
```

```python
net = NeuralNet.build([2, 10, 1], "relu")
net.train(trainX, trainY, iters = 100, alpha=.5, silent=True)
print(net.test(testX, testY))
print(net.test(trainX, trainY))
```

```python
altTrainX = np.copy(trainX)
altTrainY = np.array([[x for x in trainX[1]]])
altTestX = np.copy(testX)
altTestY = np.array([[x for x in testX[1]]])
for i in range(altTrainY.shape[1]):
    altTrainX[1][i] = trainY[0][i]
for i in range(altTestY.shape[1]):
    altTestX[1][i] = testY[0][i]
```

```python
altNet = NeuralNet.build([2,10,1], "relu")
altNet.train(trainX, trainY, iters=100, alpha=0.5, silent=True)
print(altNet.test(altTrainX, altTrainY))
```

```python
count = 0
for x, y in zip(testX[0], testY[0]):
    a = altNet.predict(np.array([[x, 1]]).T)
    b = altNet.predict(np.array([[x, 0]]).T)
    if a > b:
        if y == 1:
            count += 1
    else:
        if y == 0:
            count += 1
print(count / testY.shape[1])
```
